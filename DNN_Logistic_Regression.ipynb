```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
dataset
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
print(X_train)
print(y_train)
print(X_test)
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
print(X_train)
print(X_test)
```

```python
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=32, activation='relu'))
ann.add(tf.keras.layers.Dense(units=64, activation='relu'))
ann.add(tf.keras.layers.Dense(units=128, activation='relu'))
ann.add(tf.keras.layers.Dense(units=64, activation='relu'))
ann.add(tf.keras.layers.Dense(units=32, activation='relu'))
# Add a single dense layer for logistic regression
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid', input_dim=X_train.shape[1]))
```

```python
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
ann.fit(X_train, y_train, batch_size=32, epochs=100, verbose=1)
```

```python
# Make predictions
y_pred = ann.predict(X_test)

# Set print options for precision
np.set_printoptions(precision=2)

# Print the predicted and actual values
# print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
print(y_pred, y_test)
```

```python
import pandas as pd

# Create a DataFrame with predicted and actual values
df = pd.DataFrame({'Predicted': y_pred.flatten(), 'Actual': y_test})

# Display the DataFrame
print(df)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = ann.predict(X_test)
y_predb = (y_pred > 0.5)
```

```python
cm = confusion_matrix(y_test, y_predb)
print("Confusion Matrix:")
print(cm)
```

```python
acc = accuracy_score(y_test, y_predb)
print("Accuracy:", acc)
```
